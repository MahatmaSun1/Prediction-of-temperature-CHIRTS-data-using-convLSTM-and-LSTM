```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dropout, Embedding, TimeDistributed
# from keras.layers import Conv1D, MaxPooling2D
import warnings
import pickle
warnings.filterwarnings("ignore")
```

```python
pickle_in = open("../LSTM/Data/X_flatten.pickle","rb")
# data diload sudah dalam bentuk 1D
X = pickle.load(pickle_in)
# untuk menampilkan gambar data 1D -> 2D
plt.imshow(X[0].reshape(83,114))
```

```python
X
```

```python
df = X
data_1 = df[:-48]
data_2 = df[:-36]
data_3 = df[:-24]
data_4 = df[:-12]
data_5 = df
# data tahun 1983-2011
df=df[:-60]
```

```python
# train -> data 1983-2011
train, test = df, df[-24:]
```

```python
np.array(train).shape[0]/12
```

```python
scaler = MinMaxScaler()
scaler.fit(train)
scaler.fit(data_4)
scaler.fit(test)
train = scaler.transform(train)
data_4 = scaler.transform(data_4)
test = scaler.transform(test)
```

```python
n_input = 12
n_features = 9462
batch_size=12
# supervised data
generator = TimeseriesGenerator(train, train, length=n_input, batch_size=batch_size)
validation_sequence = TimeseriesGenerator(test, test, length=n_input, batch_size=batch_size)
```

```python
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
model = Sequential()
model.add(LSTM(40, activation='relu', unit_forget_bias='zeros', input_shape=(n_input, n_features), return_sequences=True))
model.add(LSTM(80, activation='relu', unit_forget_bias='zeros', return_sequences=True))
# model.add(Dropout(0.2))
model.add(LSTM(120, activation='relu', unit_forget_bias='zeros', return_sequences=False))
# model.add(Dropout(0.2))
model.add(Dense(128, activation='sigmoid'))
# model.add(Dropout(0.2))
model.add(Dense(200, activation='sigmoid'))
# model.add(Dropout(0.2))
model.add(Dense(200, activation='sigmoid'))
# model.add(Dropout(0.2))
model.add(Dense(n_features))
model.compile(optimizer='adagrad', loss='mae', metrics=['mae'])
```

```python
model.summary()
```

```python
model.fit_generator(generator,epochs=108)
```

```python
model.evaluate_generator(validation_sequence)
```

```python
# MAE untuk semua data
pred_list[pred_list<5]=0
pred_list_new=[]
df_asli_new=[]
no=0
for pred in pred_list:
    pred_list_new.append(list(filter(lambda x: x != 0, pred_list[no])))
    df_asli_new.append(list(filter(lambda x: x != 0, df_asli[no])))
    no+=1
    
pred_list_new=np.array(pred_list_new)
df_asli_new=np.array(df_asli_new)

mae=[]
no=0
for pred1 in pred_list_new:
    mae_ = df_asli_new[no]-pred1
    mae_ = np.average(mae_)
    mae.append(mae_)
    no+=1
```
